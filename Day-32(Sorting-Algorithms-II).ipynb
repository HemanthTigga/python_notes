# Sorting Algorithms

* Selection Sort 
* Exchange Sort 
* Bubble Sort 
* Insertion Sort 

```python
#Selection Sort
a=[12,28,11,20,8,18]
for i in range(len(a)-1):
    min=i
    for j in range(i+1,len(a)):
        if a[min]>a[j]:
            min=j
    a[i],a[min]=a[min],a[i]
    print(a)
```

```python
#Exchange Sort
a=[12,28,11,20,8,18]
for i in range(len(a)-1):
    for j in range(i+1,len(a)):
        if a[i]>a[j]:
            a[i],a[j]=a[j],a[i]
    print(a)
```

```python
# Bubble Sort
```

```python
a=[12,28,11,20,8,18]
for i in range(len(a)-1):
    for j in range(0,len(a)-1-i):
        if a[j]>a[j+1]:
            a[j],a[j+1]=a[j+1],a[j]
    print(a)
```

```python
def insertionsort(a):
    for i in range(1,len(a)):
        key=a[i]
        j=i-1
        while j>=0 and a[j]>key:
            a[j+1]=a[j]
            j-=1
        a[j+1]=key
```
