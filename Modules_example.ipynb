```python
import matplotlib.pyplot as plt #Graphics Alias name to the long name
x=[1,2,3,4,5,6,7,8,9] # list 1
y1=[1,3,5,3,1,3,5,3,1] # list 2
y2=[2,4,6,4,2,4,6,4,2] # list 3
plt.plot(x,y1,label="I") # two dot operators are there
plt.plot(x,y2,label="II") # two dot operators Nested funtion call for inbuilt function
plt.plot() # empty argument consider previous stsatements
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Example")
plt.legend() #search ?
plt.show() # 
```
